## Setup and Configuration {#setup}

```python
# Import the comprehensive test suite
from datetime import datetime
import time

from IPython.display import HTML, display
import pandas as pd
from trends_earth_api_utils import (
    TEST_USERS,
    TrendsEarthAPIClient,
    display_system_overview,
    run_comprehensive_test_suite,
)

# Configuration
API_URL = "http://localhost:5000"  # Update this for your environment

print(f"ğŸŒ Trends.Earth API URL: {API_URL}")
print(f"â° Test started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"ğŸ“‹ Available test users: {list(TEST_USERS.keys())}")

# Store test start time for duration calculation
test_start_time = time.time()
```

## Quick Health Check {#health-check}

```python
# Quick connectivity and authentication test
print("ğŸ” Performing quick health check...")

# Test basic connectivity
client = TrendsEarthAPIClient(API_URL)

try:
    # Try to connect to the API base URL
    import requests

    response = requests.get(f"{API_URL}/api/v1/healthcheck", timeout=10)
    if response.status_code == 200:
        print("âœ… API is reachable")
    else:
        print(f"âš ï¸  API returned status {response.status_code}")
except Exception:
    try:
        # Fallback - try auth endpoint
        response = requests.get(API_URL, timeout=10)
        print("âœ… API base URL is reachable")
    except Exception as e2:
        print(f"âŒ Cannot reach API: {e2}")
        print("   Please check that the API is running and accessible")

# Test authentication
print("\nğŸ” Testing authentication...")
admin_user = TEST_USERS["admin"]
login_result = client.login(admin_user["email"], admin_user["password"])

if login_result:
    print(f"âœ… Authentication successful for {admin_user['email']}")

    # Get basic system info if admin login worked
    print("\nğŸ“Š Basic system overview:")
    display_system_overview(client)

    # Logout from health check
    client.logout()
else:
    print(f"âŒ Authentication failed for {admin_user['email']}")
    print("   Please check credentials in TEST_USERS configuration")

print("\n" + "=" * 50)
```

## Full Test Execution {#full-test}

```python
# Run the comprehensive test suite
print("ğŸš€ STARTING COMPREHENSIVE API TEST SUITE")
print("=" * 60)
print(f"â° Start time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"ğŸŒ API URL: {API_URL}")
print("\nThis will test:")
print("   â€¢ Authentication & token management")
print("   â€¢ User management & administration")
print("   â€¢ Script discovery & execution")
print("   â€¢ Execution monitoring")
print("   â€¢ Rate limiting")
print("   â€¢ System & swarm monitoring")
print("   â€¢ GEE script parameter validation")
print("\n" + "=" * 60)

# Execute the comprehensive test suite
test_results = run_comprehensive_test_suite(API_URL)

print("\n" + "=" * 60)
print("ğŸ“Š COMPREHENSIVE TEST SUITE COMPLETED")
print("=" * 60)
```

## Results Analysis {#results}

```python
# Analyze and display test results
if test_results:
    print("ğŸ“ˆ DETAILED RESULTS ANALYSIS")
    print("=" * 40)

    # Create results summary table
    results_summary = []

    # Authentication results
    auth_status = "âœ… PASS" if test_results.get("authentication", False) else "âŒ FAIL"
    results_summary.append(
        {
            "Test Category": "Authentication",
            "Status": auth_status,
            "Details": "Login, token refresh, logout functionality",
        }
    )

    # User management results
    user_mgmt_status = (
        "âœ… PASS" if test_results.get("user_management", False) else "âŒ FAIL"
    )
    results_summary.append(
        {
            "Test Category": "User Management",
            "Status": user_mgmt_status,
            "Details": "User profile, admin functions, user creation",
        }
    )

    # Script availability
    scripts_count = test_results.get("scripts_available", 0)
    scripts_status = "âœ… PASS" if scripts_count > 0 else "âš ï¸  LIMITED"
    results_summary.append(
        {
            "Test Category": "Script Discovery",
            "Status": scripts_status,
            "Details": f"{scripts_count} scripts found and accessible",
        }
    )

    # Rate limiting
    rate_limit_status = (
        "âœ… PASS"
        if test_results.get("rate_limiting_working", False)
        else "âš ï¸  NOT DETECTED"
    )
    results_summary.append(
        {
            "Test Category": "Rate Limiting",
            "Status": rate_limit_status,
            "Details": "Rate limiting enforcement and recovery",
        }
    )

    # Execution monitoring
    executions_count = test_results.get("executions_found", 0)
    exec_status = "âœ… PASS" if executions_count >= 0 else "âŒ FAIL"
    results_summary.append(
        {
            "Test Category": "Execution Monitoring",
            "Status": exec_status,
            "Details": f"{executions_count} executions found, monitoring functional",
        }
    )

    # Display results table
    results_df = pd.DataFrame(results_summary)
    display(HTML(results_df.to_html(index=False, escape=False)))

    # Calculate overall score
    total_tests = len(results_summary)
    passed_tests = sum(1 for r in results_summary if "âœ… PASS" in r["Status"])
    limited_tests = sum(1 for r in results_summary if "âš ï¸" in r["Status"])
    failed_tests = sum(1 for r in results_summary if "âŒ FAIL" in r["Status"])

    print("\nğŸ¯ OVERALL TEST SCORE:")
    print(f"   Total Test Categories: {total_tests}")
    print(f"   Passed: {passed_tests} (âœ…)")
    print(f"   Limited/Warnings: {limited_tests} (âš ï¸)")
    print(f"   Failed: {failed_tests} (âŒ)")
    print(f"   Success Rate: {(passed_tests / total_tests) * 100:.1f}%")

    # Overall assessment
    if failed_tests == 0 and limited_tests <= 1:
        overall_status = "ğŸ‰ EXCELLENT - API is fully functional"
    elif failed_tests == 0:
        overall_status = "âœ… GOOD - API is working with minor limitations"
    elif failed_tests <= 2:
        overall_status = "âš ï¸  PARTIAL - API has some issues but core functions work"
    else:
        overall_status = "âŒ ISSUES - API has significant problems"

    print(f"\nğŸ† OVERALL ASSESSMENT: {overall_status}")
else:
    print("âŒ No test results available - test suite may have failed to run")

# Calculate total test duration
test_duration = time.time() - test_start_time
print(f"\nâ±ï¸  TOTAL TEST DURATION: {test_duration:.1f} seconds")
print(f"ğŸ“… Completed at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```
